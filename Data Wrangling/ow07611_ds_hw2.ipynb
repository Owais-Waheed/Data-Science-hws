```python
# This command will remove all the empty coloumns in the data.
dfc = dfc.drop(['Access Code', 'Email Address'], axis=1)

dfc.head()
```

#### Cleaning the data:
Before filling in the missing value, I will first standardize the data by fixing typos, capitalizng all letters,etc. For this I will be looking at the values and try to recognize any errors by sight.


```python
'''In q.3 'date of birth' there are two problem, on index 17 there is '2 3' and then 
on index 32 there is '1993' which can be the birth year. Since the survey was conducted in 2020 
value on index 32 should be 27
'''

dfc.at[32,'Q3'] = 27
dfc.at[17,'Q3'] = 23

# dfc['Q3']
```

```python
'''In Q.4 column we have following errors:
5.2 inches can not be a person's height so it will be a typo surely. Replacing it with 52
6l and 7O are also typos they should be replaced by 61 and 70 respectively
'''

dfc['Q4'].replace('5.2','52',inplace = True)
dfc['Q4'].replace('7O','70',inplace = True)
dfc['Q4'].replace('6l','61',inplace = True)

# dfc['Q4']
```

```python
''' In Q.5 we need to use title case for every countries and also remove unecessary white spaces.
For Standardizing data I will replace all instances of USA, US with United States
'''
dfc['Q5'].replace(['USA' , 'US' ],'United States',inplace = True)
dfc['Q5'] = dfc['Q5'].str.strip()
dfc['Q5'] = dfc['Q5'].str.title()

# dfc['Q5']
```

```python
'''In Q8 I am doing a little formatting where multiple laptops are used
'''

dfc['Q8'].replace('Microsoft/WindowsApple/MacBook' , 'Microsoft/Windows and Apple/MacBook', inplace = True)
dfc['Q8'].replace('OtherAcer/Windows' , 'Acer/Windows',inplace = True)
```

```python


dfc['Q6'].replace('Economic', 'Economics', inplace = True)
dfc['Q6'].replace(['Cse', 'Computer science engineering ', 'Computer engineering'], 'Computer Science and Engineering',inplace = True)
dfc['Q6'].replace(['Bachelor Of Engineering', 'Engineering '], ' Engineering',inplace = True)
dfc['Q6'].replace('Criminology Law And Soci', 'Criminology law and social sciences',inplace = True)
dfc['Q6'].replace('Bit', 'Business Information Technology',inplace = True)
dfc['Q6'].replace('Ece', 'Electronics And Communication Engineering',inplace = True)
dfc['Q6'].replace('Yes', np.NaN,inplace = True)
dfc['Q6'].replace('B.Tech In Cs', 'Computer Science',inplace = True)
dfc['Q6'].replace('Bs Accounting', 'Accounting',inplace = True)

dfc['Q6'] = dfc['Q6'].str.strip()
dfc['Q6'] = dfc['Q6'].str.title()

dfc['Q6']
```

### Filling missing values

after successfully formatting and standardizing the data, I am going to handle missing values by dropping irrelevant coloumns and filling the data using mean, median and mode to predict the value.

```python
# Finding rows where responses were incomplete
del_row = dfc.index[dfc['Status'] == 'Incomplete']
del_row
```

''' In the given data there are 4 incomplete responses.
Out of these 4 responses, two have completed less than 30% of the survey that is 3/13 questions and 4/13 questions.
But I won't be dropping these rows as it might lead to false results.
for example, suppose section 1 has one more student than section 2, if we drop these values than section 2 will have more students

The rest of two has completed more than 50% of the survey so I am also keeping those.
'''




In the following section, I have taken mode of the coloumns Region, city, fluency in programming language, type of laptops, Graduation date, and job status 
because it is non numeric and categorical data and there is no natural or inherent order between the entries.

```python
dfc['Region'].fillna(dfc['Region'].mode()[0], inplace=True)
dfc['City / Town / District'].fillna(dfc['City / Town / District'].mode()[0], inplace=True)


dfc['Q5'].fillna(dfc['Q5'].mode()[0], inplace=True)
dfc['Q6'].fillna(dfc['Q6'].mode()[0], inplace=True)
dfc['Q7'].fillna(dfc['Q7'].mode()[0], inplace=True)
dfc['Q8'].fillna(dfc['Q8'].mode()[0], inplace=True)
dfc['Q9'].fillna(dfc['Q9'].mode()[0], inplace=True)
dfc['Q10'].fillna(dfc['Q10'].mode()[0], inplace=True)
dfc['Q11'].fillna(dfc['Q11'].mode()[0], inplace=True)
dfc['Q12'].fillna(dfc['Q12'].mode()[0], inplace=True)



#dfc['Region'] 
#dfc['City / Town / District']
# dfc['Q5']
# dfc['Q6']
# dfc['Q7']
# dfc['Q8']
# dfc['Q9']
# dfc['Q10']
# dfc['Q11']
# dfc['Q12']
```

```python
''' IN q.4 there is height in inches so I will be taking mean to predict the value because it is numerical value and mean will not change the central tendency of the data.
'''

dfc['Q4'] = pd.to_numeric(dfc['Q4'])
dfc['Q4'].fillna(dfc['Q4'].mean(), inplace=True)

# dfc['Q4']
```
