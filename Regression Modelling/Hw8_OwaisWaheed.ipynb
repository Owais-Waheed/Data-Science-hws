```python
# Importing required libraries 

import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
```

### Q.1 Create a regression model between TotalWorkingYears and MonthlyIncome (response variable)(y). Show the scatter plot with regression line between them.

```python
X = sm.add_constant(df['TotalWorkingYears'])
Y = df['MonthlyIncome']

model = sm.OLS(Y,X).fit()
model.summary()
```

```python
plt.plot(df['TotalWorkingYears'].values, df['MonthlyIncome'].values,'ro')
plt.plot(df['TotalWorkingYears'].values ,model.fittedvalues.values,  'b')
plt.legend(['Data', 'Fitted model'])
plt.xlabel('TotalWorkingYears')
plt.ylabel('MonthlyIncome')
```

Analysis of the Relationship Between TotalWorkingYears and MonthlyIncome

Y (predivtion) = MonthlyIncome
X = Total working years 
R-squared (R²) =0.597
P-value (Prob (F-statistic))= 2.73e-292
The linear regression model shows a strong and statistically significant positive relationship between TotalWorkingYears and MonthlyIncome.

The p-value being less than 0.05 indicates a significant linear connection between TotalWorkingYears and MonthlyIncome. As TotalWorkingYears increases, MonthlyIncome also goes up. The lower the p-value, the stronger the relationship. This strong connection is evident in the graph where data points closely align with the regression line.

This equation means that for every additional year an employee works, their monthly income increases by $468.

R-squared Value Interpretation:

R-squared (R²): 0.597
The R-squared value of 0.597 indicates that approximately 59.7% of the variance in MonthlyIncome can be explained by the variation in TotalWorkingYears. This suggests that the model is a good fit for the data.

In summary, the data supports a strong positive relationship between TotalWorkingYears and MonthlyIncome, and the linear regression model provides a good fit for explaining the variation in MonthlyIncome based on TotalWorkingYears. This finding aligns with the common observation that, in many companies worldwide, as employees accumulate more years of work experience, their monthly income tends to increase.

Moreover, Linear Equation for the Relationship:

The linear equation for this relationship can be expressed as follows:
y = 1227.935 + 467.65x

It means that every year the monthly income increases by 467.65 units .

The p-value associated with the F-statistic is high, at 0.948. A high p-value indicates that the model is not statistically significant. In this case, the model does not explain the variance in DistanceFromHome well.

In summary, based on the regression results, it appears that Age is not a significant predictor of an employee's distance from home (DistanceFromHome). The low R-squared value, high p-value, and other statistical tests indicate that this model does not provide a good fit for explaining the variation in DistanceFromHome based on Age

### Q4.Create a regression model to predict MonthlyIncome using all other inputs. Discuss the effectiveness of the model. Report 3 most significant inputs and 3 least significant inputs (based on p-value) and interpret the results. Create a new input and show the prediction of MonthlyIncome using the same model

```python
# Using one hot encoding technique to make categorical data suitale for machine learning model
categoricalcoloumns = df.select_dtypes(exclude=['int','float']).columns.tolist()
df4 = pd.get_dummies(df, columns=[x for x in categoricalcoloumns])

X = df4.drop(['MonthlyIncome'],axis=1)
X = sm.add_constant(X)
Y = df4['MonthlyIncome']

modelall = sm.OLS(Y,X).fit()
modelall.summary()
```

##### Effectiveness Of model

R-squared (R²): 0.944

The R-squared value of 0.944 is relatively high, indicating that the model explains a significant portion of the variance in MonthlyIncome. This suggests that the model is quite effective at predicting MonthlyIncome.

Adjusted R-squared (Adj. R²) = 0.943

The adjusted R-squared is also high, suggesting that adding more independent variables to the model hasn't caused overfitting.

Prob (F-statistic): 0.00

The p-value associated with the F-statistic is very close to 0, further supporting the model's overall significance.

##### Significant Inputs (Low P-values):

JobLevel: p < 0.001 <br>
TotalWorkingYears: p < 0.001 <br>
JobRole (all subfields) : p < 0.001 <br>
These are the three most significant inputs. They have low p-values, suggesting a strong impact on MonthlyIncome.

##### Least Significant Inputs (High P-values):

Years In current role: p = 0.908 <br>
TrainingTimesLastYear: p = 0.469 <br>
WorkLifeBalance: p = 0.588

These are the three least significant inputs. Their high p-values suggest a weak impact on MonthlyIncome.

##### Interpretation of Results:

The model is highly effective, with a high R-squared value, indicating that it explains a large portion of the variance in MonthlyIncome.
JobLevel, TotalWorkingYears, and the job role are the most significant factors affecting MonthlyIncome, with very low p-values.
Education, TrainingTimesLastYear, and WorkLifeBalance have high p-values, indicating that they are not statistically significant predictors of MonthlyIncome.
The model also accounts for categorical variables like job roles and marital status (because of hot encoding), showing their impact on MonthlyIncome.


```python
#Regression model
new_input = pd.DataFrame({
    'Age': [56],
    'Department': ['Research & Development'],
    'DistanceFromHome': [15],
    'Education': [3],
    'EducationField': ['Life Sciences'],
    'Gender': ['Male'],
    'HourlyRate': [60],
    'JobLevel': [2],
    'JobRole': ['Laboratory Technician'],
    'MaritalStatus': ['Single'],
    'NumCompaniesWorked': [6],
    'OverTime': ['No'],
    'PerformanceRating': [3],
    'StockOptionLevel': [1],
    'TotalWorkingYears': [0],
    'TrainingTimesLastYear': [3],
    'WorkLifeBalance': [3],
    'YearsAtCompany': [4],
    'YearsInCurrentRole': [2],
    'YearsSinceLastPromotion': [1],
    'YearsWithCurrManager': [2]
})

#Concate new input with original data
new_input = pd.concat([df, new_input], ignore_index=True)
#drop MonthlyIncome column
new_input = new_input.drop(['MonthlyIncome'], axis=1)
#get dummies
df4b = pd.get_dummies(new_input, columns=[x for x in categoricalcoloumns])
#add constant
df4b = sm.add_constant(df4b)
#predict
new_input = df4b.tail(1)
print("The predicted MonthlyIncome is: ")
modelall.predict(new_input)
```

According to our prediction model, the predicted monthly income of the new employee (depending on the new input values in the variable) will be 4848 units

##### Interpretation of Results:
